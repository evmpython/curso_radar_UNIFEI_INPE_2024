# **4 Passo:** Declarando funes

```python
#-----------------------------------------------------------------------------------
#  Funo que calcula a latitude e longitude do azimute
#-----------------------------------------------------------------------------------
def calc_latlon_azimute(latitude_radar, longitude_radar, distancia_radar, azimute_escolhido):

    """
    Calcula a latitude e longitude de um segundo ponto no espao

    Parmetros de entrada:
                           latitude_radar (float): latitude do radar em graus
                           longitude_radar (float): longitude do radar em graus
                           distancia_radar (float)): distncia do radar ao segundo ponto em km
                           azimute_escolhido (float): ngulo azimute em graus

    Parmetros de sada:
                        lat_azimute (float): latitude do segundo ponto em graus
                        lon_azimute (float): longitude do segundo ponto em graus

    Exemplo:
             calc_latlon_azimute(-23.600795, -45.97279, 250, 120)

    """

    # importa funes
    import geopy
    from geopy import distance

    # given: lat1, lon1, b = bearing in degrees e d = distance in kilometers
    ponto_partida = geopy.Point(latitude_radar, longitude_radar)

    # calcula latitude e longitude do ponto 2
    destino = distance.distance(kilometers=distancia_radar).destination(ponto_partida, azimute_escolhido)
    lat_azimute, lon_azimute = destino.latitude, destino.longitude

    return lat_azimute, lon_azimute

#-----------------------------------------------------------------------------------
#  Funo que plota circulos de distncia em geral
#-----------------------------------------------------------------------------------
def evm_plota_aneis_em_geral(aneis, lon_r, lat_r, color, label):

    """
    Retorna crculos de distncia centrado no radar

    Parmetros de entrada:
                           aneis (lista): tamanho do raio do crculo em km
                           lon_r (float): valor da longitude do centro do crculo em graus
                           lat_r (float): valor da latitude do centro do crculo em graus
                           color (str): cor do raio do crculo
                           label (str): legenda

    Parmetros de sada:
                         latitude e longitude que delimitam os crculos de distncia centrado no radar e plota os crculos

    Exemplo:
             evm_plota_aneis_em_geral([100], -45.97279, -23.600795, 'gray', label='Radar: 100 km')
    """

    origin = geopy.Point(lat_r, lon_r)

    lons = np.zeros((len(aneis), 361))
    lats = np.zeros((len(aneis), 361))
    for i, dis in enumerate(aneis):
        xpts = []
        ypts = []
        for az in range(361):
            destination = distance.distance(kilometers=dis).destination(origin, az)
            lat2, lon2 = destination.latitude, destination.longitude
            xpts.append(lon2)
            ypts.append(lat2)
        lons[i,:] = xpts[:]
        lats[i,:] = ypts[:]

    for i, anel in enumerate(aneis):
        ax.plot(lons[i,:], lats[i,:], color=color, label= label)
```

```python
#========================================================================================================================#
#                                                LEITURA DADO DO RADAR
#========================================================================================================================#
# leitura do arquivo
filename = '2024032118000300dBZ.vol.h5'
radar = pyart.aux_io.read_odim_h5(f'/content/{filename}')

# extrai a latitude e longitude do radar
lat_radar, lon_radar = radar.latitude['data'][0], radar.longitude['data'][0]

# limites dos dados
latmin, latmax = radar.gate_latitude['data'].min(), radar.gate_latitude['data'].max()
lonmin, lonmax = radar.gate_longitude['data'].min(), radar.gate_longitude['data'].max()

# extrai a data do dado do radar
data = pyart.util.datetime_from_grid(radar)

#========================================================================================================================#
#                                                SEO TRANSVERSAL
#========================================================================================================================#
# escolha o ngulo azimute para fazer a seo transversal
azi = 120

# latitude e longitude do azimute escolhido
lat_azi, lon_azi = calc_latlon_azimute(lat_radar, lon_radar, 250, azi)

# funo que produz a seo transversal
cross_section = pyart.util.cross_section_ppi(radar, [azi])

#========================================================================================================================#
#                                                PLOTA FIGURA
#========================================================================================================================#
# cria moldura da figura
fig = plt.figure(figsize=[13,5])

#==========================#
#     FIGURA 1: PPI
#==========================#
# moldura da figura
ax1 = plt.subplot(121, projection=ccrs.PlateCarree())

# define a elevao
level = 2

# extrai o ngulo de elevao da primeira elevao (level=0)
str_level = str(radar.fixed_angle["data"][level])

# cria o display
display = pyart.graph.RadarMapDisplay(radar)

# plota o mapa
display.plot_ppi_map('reflectivity_horizontal',
                     sweep=level,
                     vmin=20,
                     vmax=70,
                     ax=ax1,
                     lat_0=lat_radar,
                     lon_0=lon_radar,
                     projection=ccrs.PlateCarree(central_longitude=lon_radar),
                     resolution='10m',
                     cmap='pyart_NWSRef',
                     colorbar_label='Refletividade Horizontal (dBZ)',
                     mask_outside=False)

# plota as linhas transversais que passam pelo centro do radar
display.plot_line_geo(np.array([lon_radar, lon_radar]), np.array([latmin, latmax]), linestyle='--', color='black')
display.plot_line_geo(np.array([lonmin, lonmax]), np.array([lat_radar, lat_radar]), linestyle='--', color='black')

# plota aneis de distncia
display.plot_range_rings([100, 200, 250], ls='--', lw=1, col='gray', ax=ax1)

# indica a localizao do radar com um ponto
display.plot_point(lon_radar, lat_radar, symbol='o', label_text='Radar', color='red')

# plota a linha do ngulo azimute
display.plot_line_geo(np.array([lon_radar, lon_azi]), np.array([lat_radar, lat_azi]), linestyle='--', color='red', lw=2)

# plota as linhas de latitude e longitudes
ax = plt.gca()
gl = ax.gridlines(crs=ccrs.PlateCarree(), draw_labels=True, linewidth=1, color='gray', alpha=0.5, linestyle='--')
gl.xlabels_top = False
gl.ylabels_right = False

# ttulo da figura
plt.title(f'PPI {str_level}$\degree$: {data} UTC', fontsize=15)
plt.title('a)', loc='left', fontsize=15)

#==========================#
#     FIGURA 2: CORTE
#==========================#
# moldura da figura
ax2 = plt.subplot(122)

# monta um objeto "display" do Py-ART
display = pyart.graph.RadarDisplay(cross_section)

# faz o plot
display.plot('reflectivity_horizontal',
             vmin=20,
             vmax=70,
             cmap='pyart_NWSRef',
             colorbar_label='Refletividade Horizontal (dBZ)',
             ax=ax2)

# seta os limites dos eixos x e y
display.set_limits(ylim=[0,15])
display.set_limits(xlim=[0,250])

# nome dos eixos x e y
plt.xlabel('Distncia (km)')
plt.ylabel('Altura (km)')

# ttulo figura
plt.title(f'Seo Vertical {azi}$\degree$', fontsize=15)
plt.title('b)', loc='left', fontsize=15)

# recorta figura
plt.tight_layout()

# salva figura
plt.savefig(f'/content/Atividade_2_exercicio_a.png',
             bbox_inches='tight',
             dpi=300)
```

# **EXERCCIO B):** Produzindo e plotando Constant Altitude Plan Position Indicator (CAPPI)


---

**ENUNCIADO:** O objetivo deste exerccio  o aluno produzir e plotar o CAPPI de 4000 m de altitude da refletividade horizontal. Para isto vocs utilizaro como referncia o `Script 07` da aula prtica. Para realizar esse exerccio vocs  devero fazer o seguinte passo-a-passo:

1.   Leitura do arquivo do radar. Insira o nome do arquivo que foi baixado. Por exemplo: `filename = '2024032118000300dBZ.vol.h5'`. Vocs devero inserir o nome do arquivo que foi baixado e feito o upload no `/content/`, e que provavelmente ser de uma data e horrio diferente desse exemplo.

2.   Na funo que realiza o clculo do CAPPI `cappi = pyart.map.grid_from_radar` usar as seguintes definies:
     - dimenses da grade: `grid_shape=(2, 500, 500)`
     - limites da grade: `grid_limits=(
        (3000-altitude, 4000-altitude), (-250000, 250000), (-250000, 250000))`

3. Em `display.plot_grid` defina as seguintes propriedades da figura:
     - varivel: `'reflectivity_horizontal'`
     - nvel do CAPPI: `level=1`
     - valor mnimo da refletividade: `vmin=20`
     - valor mximo da refletividade: `vmax=70`
     - nome da paleta de cores: `cmap='pyart_NWSRef`
     - nome na barra de cores: `colorbar_label='Refletividade Horizontal (dBZ)`
  
4. Em `display.plot_range_rings` plotar os aneis de 100, 200 e 250 km.

5. No ttulo da figura: `plt.title(f'CAPPI {int(display.grid.z["data"][1]) + altitude} m - Refletividade: {data}', fontsize=10)`

5. Mostrar a figura na tela.

6. Salvar a figura em PNG com 300 dpi de resoluo e com o nome `Atividade_2_exercicio_b.png`.

`*OBSERVAO:` Dependendo do arquivo que foi escolhido pode ser que naquela imagem no tenha atividade convectiva. Nessa situao pode ser que a figura fique sem informao ou na cor branca.

```python
#========================================================================================================================#
#                                             LEITURA DO DADO DE RADAR
#========================================================================================================================#
# leitura do arquivo
filename = '2024032118000300dBZ.vol.h5'
radar = pyart.aux_io.read_odim_h5(f'/content/{filename}')

# extrai a latitude, longitude do radar e altitude do radar
lat_radar, lon_radar = radar.latitude['data'][0], radar.longitude['data'][0]
altitude = radar.altitude['data'][0]

# limites dos dados
latmin, latmax = radar.gate_latitude['data'].min(), radar.gate_latitude['data'].max()
lonmin, lonmax = radar.gate_longitude['data'].min(), radar.gate_longitude['data'].max()

# extrai a data do dado do radar
data = pyart.util.datetime_from_grid(radar)

#========================================================================================================================#
#                                             GERA O CAPPI
#========================================================================================================================#
# gera CAPPI
cappi = pyart.map.grid_from_radars(radar,
                                   grid_shape=(2, 500, 500),
                                   grid_limits=((3000. - altitude, 4000. - altitude),
                                                (-250000., 250000.),
                                                (-250000., 250000.)),
                                   grid_origin=(lat_radar, lon_radar),
                                   gridding_algo='map_gates_to_grid',
                                   roi_func='dist_beam',
                                   min_radius=1750.0,
                                   weighting_function='Nearest',
                                   fields=['reflectivity_horizontal'])

#========================================================================================================================#
#                                                     PLOTA FIGURA
#========================================================================================================================#
# define o tamanho da figura
fig = plt.figure(figsize=(10,7))

# define os eixos e projeo da figura
ax = fig.add_subplot(111, projection=ccrs.PlateCarree())

# monta um objeto "display" do Py-ART
display = pyart.graph.GridMapDisplay(cappi)

# plota o cappi
display.plot_grid("reflectivity_horizontal",
                  level=1,
                  vmin=40,
                  vmax=70,
                  ax=ax,
                  cmap='pyart_NWSRef',
                  colorbar_label='Refletividade Horizontal (dBZ)')

# plotas as linhas passando pelo radar
display.plot_crosshairs(lon=lon_radar, lat=lat_radar)

# plota os aneis de distncia do radar
evm_plota_aneis_em_geral([100, 200, 250], lon_radar, lat_radar, 'black', label='')

# ttulo da figura
plt.title(f'CAPPI {int(display.grid.z["data"][1]) + altitude} m - Refletividade: {data}', fontsize=10)

# nome dos eixos x e y
ax.set_xlabel('Longitude ($\degree$)', fontsize=15)
ax.set_ylabel('Latitude ($\degree$)', fontsize=15)

# recorta figura
plt.tight_layout()

# salva figura
plt.savefig(f'/content/Atividade_2_exercicio_b.png', dpi=300)
```
